```python
class Book:
    def __init__(self, code, title, author, publisher):
        self.code = code
        self.title = title
        self.author = author
        self.publisher = publisher
        
    def book_info(self):
        print(f'<도서 {self.title}의 상세정보>')
        print(f'코드: {self.code}')
        print(f'저자: {self.author}')
        print(f'출판사: {self.publisher}')
        print()
        
def set_info():   
    code = input('코드를 입력하세요: ')
    title = input('제목을 입력하세요: ')
    author = input('저자를 입력하세요: ')
    publisher = input('출판사를 입력하세요: ')
    print()
# Book클래스를 통해 book이라는 객체를 만들어 입력받은 정보를 담는다
    book = Book(code, title, author, publisher) 
    return book
    
    
def print_info_all(book_list):
    for book in book_list:
        book.book_info()
        
def print_info(book_list,title):
    for idx, book in enumerate(book_list):
        if book.title == title:
            book_list[idx].book_info() 
# 리스트안 요소는 객체, 객체의 메소드에 접근하고 싶을 때는 위와 같이 .메소드로 접근한다
# print(book_list[idx])는 객체의 메모리 위치만 가져올 뿐!!!!!!!!

def delete_book(book_list, title):
    for idx, book in enumerate(book_list):
        if book.title == title:
            del book_list[idx]

def repo_book(book_list):
    with open('./book_db.txt','a',encoding='utf8') as f:
        for book in book_list:
            f.write(book.code + '\n')
            f.write(book.title + '\n')
            f.write(book.author + '\n')
            f.write(book.publisher + '\n') 
            # f.write()는 하나의 인수만 받는다
            
def load_book(book_list):
    with open('./book_db.txt','r',encoding='utf8') as f:
        lines = f.readlines()
        num = int(len(lines)/4) # 책이 몇권인지 알 수 있음
        # 책 1권의 정보는 4줄, 8줄이면 2권이 있는 확인할 수 있다
       
        # 따라서 다음 num에 2를 넘겨주면 
        # i = 0인 경우 리스트 lines의 인덱스 요소값 0~3까지 각 변수로 들어가고
        # 이 변수들을 Book 클래스의 객체 book에 담는다, 그 후 북리스트에 객체 추가
        # i = 1인 경우 리스트 lines의 인덱스 요소값 4~7까지 각 변수로 들어가고
        # 이 변수들을 Book 클래스의 객체 book에 담는다, 그 후 북리스트에 객체 추가
        # 즉, lines[4*i] 인 이유는 책의 정보가 총 4줄이기 때문에 4줄씩 끊어오기 위함
      
        for i in range(num):
            code = lines[4*i].strip()
            title = lines[4*i+1].strip()
            author = lines[4*i+2].strip()
            publisher = lines[4*i+3].strip()
            book = Book(code, title, author, publisher)
            book_list.append(book)
        

def menu_display():
    menu = """
-------------------
1. 도서 정보 입력
2. 도서 정보 출력
3. 도서 삭제
4. 전체 도서 정보 출력
5. 종료
-------------------
"""
    print(menu)
    menu = int(input('메뉴를 선택하세요: '))
    return menu

        
def run():
    book_list = []
#     load_book(book_list)
    
    while True: 
        menu_num =  menu_display()
        
        if menu_num == 1:
            book = set_info()
            book_list.append(book)
        
        elif menu_num == 2:
#             print_info(book_list)
            title = input("출력하고자 하는 도서명: \n")
            print_info(book_list,title)   
            
        elif menu_num == 3:
            title = input("삭제하고자 하는 도서명: \n")
            delete_book(book_list, title)
        
        elif menu_num == 4:
            print_info_all(book_list)
            
        elif menu_num == 5:
            repo_book(book_list)
            print("프로그램 종료")
            break
            
#     print_info(book_list)
    
        
if __name__ == '__main__':
    run()
```
