```python
import requests

# 세션 활성화하기
sess = requests.Session()

res = sess.get('https://www.naver.com')

# 응답된 데이터(텍스트로 가져올 수 있다)
print(res.text) # 해당 url의 html코드를 가져옴

# 상태코드 확인
print(f'Status Code : {res.status_code}') # 200
print(f'Okay : {res.ok}') # True


# 세션 비활성화(종료)
sess.close()
```

```python
import requests

ss2 = requests.Session()

# httpbin.org에 쿠키를 만들어서 전송하여 return 받기
re = ss2.get('http://httpbin.org/cookies', cookies={'name': 'kim'})
print(re.text)

# user-agent
url = 'http://httpbin.org/get'
headers = {'user-agent':'test_browser'}

re2 = ss2.get(url, headers=headers)
print(re2.text)

ss2.close()
```
