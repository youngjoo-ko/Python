```python
import urllib.parse as rep
import urllib.request as req
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
import os  # 운영체제

# 헤더 정보 초기화해서 요청하기(일반적인 형식)
opener = req.build_opener()
opener.addheaders = [('User-agent', UserAgent().ie)]
req.install_opener(opener) # 위에서 세팅한 header 정보대로 설정

base_url = 'https://search.naver.com/search.naver?where=image&query='

# url 인코딩을 위한 설정
quote = rep.quote_plus('고양이') # quote 는 관례적인 변수
url = base_url + quote
# print(f'request URL : {url}')

res = req.urlopen(url)
save_path = 'c:/python_images/'
# 폴더 유무에 따른 예외처리(문제 발생 시 프로그램 종료)
try: 
    if not(os.path.isdir(save_path)):  # images 폴더가 없으면
        os.makedirs(os.path.join(save_path)) # 새로 만들어라
    
except:
    print('폴더 생성 실패')
    raise RuntimeError('system 오류')
else:
    print('폴더 생성 완료')


bs = BeautifulSoup(res, 'html.parser')
# print(bs.prettify())

# select 사용해서 이미지 가져오기
img_list = bs.select('div.img_area._item > a.thumb._thumb > img')
print(img_list)

for idx, img in enumerate(img_list, 1): # 인덱스 1번부터 불러오겠다
    # 이미지 url 확인
#     print(idx, img['data-source'])
    file_name = os.path.join(save_path, str(idx) + '.png')
    req.urlretrieve(img['data-source'], file_name)
print('다운로드 성공')
```
