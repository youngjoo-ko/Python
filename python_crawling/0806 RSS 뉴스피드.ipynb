```python
# RSS : 사이트에서 새로운 소식을 제공하는 기술

import urllib.request
import urllib.parse

rssurl = 'https://news.sbs.co.kr/news/SectionRssFeed.do'

params = []

for query in ['01', '02', '07']:
    params.append(dict(sectionID=query))
    
print(params)

for c in params:
    param = urllib.parse.urlencode(c)
    print(param)
    
    # url 만들기
    url = rssurl + '?' + param + '&plink=RSSREADER'
    print('url =', url)
    res_data = urllib.request.urlopen(url).read()
    
    # 응답 데이터를 디코딩
    contents = res_data.decode('utf-8')
    
    # 출력
    print('#'*100)
    print(contents)
```
