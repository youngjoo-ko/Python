```python
from bs4 import BeautifulSoup

html="""
<!DOCTYPE html>
<html lang="en">
<head>
    <title>Document</title>
</head>
<body>
    <h1>제목1</h1>
    <h2>제목2</h2>
    <p class='title'><strong>옛날 동화</strong></p>
    <p class="story">아주 옛날 이야기...
        <a href="http://test.com" class='sister' id='link1'>말똥이</a>
        <a href="http://test.com" class='sister' id='link2'>개똥이</a>
        <a href="http://test.com" class='sister3' id='link3'>길동이</a>
        <a href="http://test.com" data2='brother' class='sister3' id='link3'>길동이</a>
</p>
        <p class="story">이야기 끝</p>
</body>
</html>
"""

soup = BeautifulSoup(html, 'html.parser') 
# 첫 인자는 html소스코드, 두 번째 인자는 어떤 parser를 이용할지 명시.

print(type(soup))

# prettify : html 코드를 보기좋게 정리해줌 
print('prettify: ', soup.prettify())

# h1 태그에 엑세스
h1 = soup.html.body.h1
print('h1 태그: ', h1)

# p 태그에 엑세스
p1 = soup.html.body.p
print('p 태그: ', p1)

p2 = p1.next_sibling.next_sibling #p1태그의 </p> 태그로 끝난 다음에 커서가 놓인다고 생각하자
print('p2 태그: ', p2)

# string : 컨텐트(텍스트) 출력
print('h1 태그의 텍스트: ', h1.string)
print('p1 태그의 텍스트: ', p1.string)

print(p2.next_element) #  p2에 놓인 커서의 다음 내용을 출력한다

# find, find_all 사용
# # soup에 있는 a태그를 모두 가져오기
link1 = soup.find_all('a')
print(link1) 

# 처음 a태그만 가져오기
link2 = soup.find('a') 
print(link2)

# class 이름이 sister인 a 태그 모두 가져오기(이때, 변수에 class_ 로 쓰는것 유의!)
link3 = soup.find_all('a', class_='sister')
print(link3)

# class 이름이 sister인 첫번쨰 a 태그 가져오기
link4 = soup.find('a', class_='sister') 
print(link4)

# class 이름이 sister, data2 속성이 brother인 a 태그 가져오기
link5 = soup.find('a', {'class':'sister3', 'data2':'brother'}) 
print(link5)

# select(find_all과 비슷), select_one(find)
link6 = soup.select_one('p.title > strong')
print(link6)

# a태그면서 id가 link2인 것 가져오기
link7 = soup.select_one('a#link2') # 아이디는 #으로 접근
print(link7) 
print(link7.string)

# p태그의 story 클래스 모두 가져오기
link8 = soup.select('p.story')
print(link8)

# 클래스네임이 story인 p태그의 자식 태그 중 a 태그 모두 가져오기
link9 = soup.select('p.story > a') # 결과를 리스트로 리턴해줌
print(link9) 

# p.story 밑에 두번쨰 a 태그 가져오기(가상 클래스 이용)
link10 = soup.select_one('p.story > a:nth-of-type(2)')
print(link10)
```
