```python
import json
import requests

ss = requests.Session()

# httpbin.org 사이트에 10개의 json 데이터를 요청해보기
res = ss.get('http://httpbin.org/stream/10')

# 수신 데이터 확인
print(res.text)

# encoding 확인하기
print(f'encoding:{res.encoding}')

if res.encoding is None:
    res.encoding = 'UTF-8' 
    print(f'encoding:{res.encoding}')
    
# 수신 데이터를 DICT 타입으로 변환
for line in res.iter_lines(decode_unicode=True):
    print(line)
    print(type(line))
    
    #json(dict) 변환 후 타입 확인하기
    b = json.loads(line)
    print(type(b))
    
    for keys, values in b.items():
        print(f'key:{keys}, value:{values}')
        
ss.close()
```
